# Random Forest Classifier

This notebook demonstrates how to use a Random Forest Classifier in Concrete ML. We will train a `scikit-learn` Random Forest model and then convert it to its FHE equivalent using Concrete ML.

```
import time

import numpy
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier as SklearnRandomForestClassifier
from sklearn.model_selection import train_test_split

from concrete.ml.sklearn import RandomForestClassifier

# Generate a synthetic dataset
x, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=0, random_state=42)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Train a scikit-learn Random Forest Classifier

```
sklearn_model = SklearnRandomForestClassifier(n_estimators=10, max_depth=5, random_state=42)
sklearn_model.fit(x_train, y_train)
```

### Convert the scikit-learn model to a Concrete ML model

```
# Convert the sklearn model to a Concrete ML model
concrete_model = RandomForestClassifier.from_sklearn_model(sklearn_model, n_bits=6)
```

### Compile the Concrete ML model

```
# Compile the model
circuit = concrete_model.compile(x_train)
```

### Perform FHE inference and compare results

```
# Run the inference in FHE
y_pred_fhe = concrete_model.predict(x_test, fhe="execute")

# Get the clear predictions from the original sklearn model
y_pred_clear = sklearn_model.predict(x_test)

# Compare the results
accuracy = numpy.mean(y_pred_fhe == y_pred_clear)
print(f"Accuracy of FHE predictions vs clear predictions: {accuracy * 100:.2f}%")
```
